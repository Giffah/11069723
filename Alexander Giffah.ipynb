```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Dataset = pd.read_csv("red_wine.csv")
```

```python
plt.figure(figsize=(8, 6))
plt.hist(Dataset['alcohol'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Alcohol Content in Red Wine')
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
# plt.show()
plt.savefig("plot1.png")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(Dataset['alcohol'], Dataset['pH'], alpha=0.5, color='green')
plt.title('Alcohol Content vs. pH in Red Wine')
plt.xlabel('Alcohol Content')
plt.ylabel('pH')
# plt.show()
plt.savefig("plot2.png")
```

```python
corr_matrix = Dataset.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
plt.savefig("plot3.png")
```

```python
numerical_columns = Dataset.select_dtypes(include='number').columns
num_columns = len(numerical_columns)
rows = (num_columns // 3) + (num_columns % 3)
cols = 3
plt.figure(figsize=(15,10))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(rows, cols, i)
    sns.violinplot(x=column, data=Dataset, color='skyblue')
    plt.title(f'Violin Plot of {column}')
plt.tight_layout()
plt.savefig("Plot4.png",dpi = 300)
```
